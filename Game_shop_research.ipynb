У нас есть значение `tbd`, могу предположить, что это сокращение рассшифровывается как - `To Be Determined`, то есть "Будет определено", но сказать точно я не могу, т.к. вводных в тз, по этому поводу, я не увидел, буду рассматривать `tbd` как пропуски.

```python
# Преобразуем неявные пропуски `tbd` в NaN и изменю тип данных на float
tbd = df['user_score'] != 'tbd'
df['user_score'].where(tbd, np.nan, inplace=True)
df["user_score"] = df["user_score"].astype('float64')
df["critic_score"] = df["critic_score"].astype('Int64') # Сразу изменю тип данных в столбце critic_score на int

print(len(df[df["user_score"] == "tbd"]))
```

#### Год выпуска

269 пропусков из 16 715 записей – это ~1.6% датасета. Даже если мы не сумеем заполнить эти пропуски, ничего критичного не случится.

Попробую заполнить пропуски исходя из даты выпуска игры на другой платформе. Конечно, может случится и так, что игра для какой-то плафтормы выйдет позже, но в среднем издатели стараются выпускать мультиплатформенные проекты в один год. ## Шаг 5. Проверка гипотез
Для каждой из приведенных ниже гипотез проверим две (двусторонние) гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам. Чтобы понять, различаются ли пользовательские рейтинги и значима ли разница между средними значениями, проведем статистические тесты.

<a id="cell6"></a>
### 5.1 Средние пользовательские рейтинги платформ Xbox One и PC одинаковые
Мы рассматриваем две генеральные сопокупности – игры на разных платформах.

$H_0$ - Средние пользовательские рейтинги платформ Xbox One (`XOne`) и `PC` `одинаковые`.

$H_1$ - Средние пользовательские рейтинги платформ Xbox One (`XOne`) и `PC` `различаются`.

$\alpha = 1\%$ - из-за Большого количества пропусков в датасете.

Буду использовать метод `st.ttest_ind()` для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам;

Посмотрим и сравним дисперсию двух выборок, если она отличаеться более чем на 10% тогда в Т-тесте зададим параметр, `equal_var` = `False`(по умолчанию `True`) и дисперсия будет оценена для каждый выборки по отдельности, а не общей.

```python
xone = df.query("platform == 'XOne'")
pc = df.query("platform == 'PC'")

print(f"Среднее XOne: {xone['user_score'].mean()}\nCреднее PC: {pc['user_score'].mean()}\n")
print(f"Дисперсиия XOne: {xone['user_score'].var()}\nДисперсия PC: {pc['user_score'].var()}")
```

Мы видим что диспеерсии выборок отличается более чем на 10%, берем аргумент `equal_var`=`False` в Т-тесте.

```python
result = st.ttest_ind(xone["user_score"],
                      pc["user_score"],
                      equal_var=False,
                      nan_policy='omit' # Игнорирование пропусков
                     )

alpha=0.01

print(f"p-значение: {result.pvalue}")

if (result.pvalue < alpha):
    print("Отвергаем нулевую гипотезу.")
else:
    print("Не получилось отвергнуть нулевую гипотезу.")
```

У нас не получилось опровергнуть нулевую гипотезу о равнстве генеральных совокупностей с $\alpha = 1\%$, соответственно полученное значение `p-value` говорит о том, что такое или большее различие можно получить случайно с вероятностью 0.13%.

<a id="cell7"></a>
### 5.2 Средние пользовательские рейтинги жанров Action и Sports разные
Мы рассматриваем две генеральные сопокупности – игры разных жанров.

$H_0$ - Средние пользовательские рейтинги жанров `Action` и `Sports` **одинаковые**.

$H_1$ - Средние пользовательские рейтинги жанров `Action` и `Sports` **различаются**.

$\alpha = 5\%$

Буду использовать метод `st.ttest_ind()` для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам;

Посмотрим и сравним дисперсию двух выборок, если она отличаеться более чем на 10% тогда в Т-тесте зададим параметр, `equal_var` = `False`(по умолчанию `True`) и дисперсия будет оценена для каждый выборки по отдельности, а не общей.

```python
action = df.query("genre == 'Action'")
sports = df.query("genre == 'Sports'")

print(f"Среднее Action: {action['user_score'].mean()}\nCреднее Sports: {sports['user_score'].mean()}\n")
print(f"Дисперсиия Action: {action['user_score'].var()}\nДисперсия Sports: {sports['user_score'].var()}")
```

Мы видим что диспеерсии выборок отличается более чем на 10%, берем аргумент `equal_var`=False в Т-тесте.

```python
results = st.ttest_ind(action['user_score'],
                       sports['user_score'],
                       equal_var=False,
                       nan_policy='omit' # Игнорирование пропусков
                      )

alpha = 0.05

print(f"p-значение: {results.pvalue}")

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

У нас получилось опровергнуть нулевую гипотезу о равнстве генеральных совокупностей с $\alpha = 5\%$, соответственно полученное значение `p-value` говорит о том, что такое или большее различие можно получить случайно с **очень маленькой** вероятностью .
